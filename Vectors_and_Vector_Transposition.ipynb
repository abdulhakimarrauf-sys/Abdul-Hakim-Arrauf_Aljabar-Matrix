```python
# buat vector x dengan ukuran 3x1
import numpy as np

x = np.array([[19],
              [20],
              [10]])
x
```

```python
# tampilkan ukuran vector x
x.shape
```

```python
# tampilkan tipe data dari vector x
type(x)
```

```python
"""**SOAL NO 2 (VECTORS AND VECTOR TRANSPOSITION)**"""
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
x_trans = x.T
x_trans
```

```python
# tampilkan ukuran dari vector x_t
x_trans.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[19],
              [20],
              [10]])
y
```

```python
# tampilkan ukuran dari vector x_t
y.shape
```

```python
"""**SOAL NO 3 (VECTORS AND VECTOR TRANSPOSITION) (TENSORFLOW)**"""
```

```python
# lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import tensorflow as tf

x_tf = tf.Variable([[19],
                    [20],
                    [10]])
x_tf

x_tf.shape

y_tf = tf.Variable([[19],
                    [20],
                    [10]])
y_tf

y_tf.shape
```

```python
"""**SOAL NO 3 (VECTORS AND VECTOR TRANSPOSITION) (PYTORCH)**"""
```

```python
# lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import torch

x_pt = torch.tensor([[19],
                     [20],
                     [10]])
x_pt

x_pt.shape

y_pt = torch.tensor([[19],
                     [20],
                     [10]])
y_pt

y_pt.shape
```
