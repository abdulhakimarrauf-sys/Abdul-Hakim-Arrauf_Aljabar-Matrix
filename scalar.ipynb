```python
"""**NO 2 (SCALAR) (TENSORFLOW)**"""
```

```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow
import tensorflow as tf
```

```python
x_tf = tf.Variable(19)
```

```python
# tampilkan tipe data dari x_tf (Tensorflow)
print(f'type dari x_tf: {type(x_tf)}')
```

```python
# buat variabel skalar y_tf dan simpan suatu nilai skalar (Tensorflow)
y_tf = tf.Variable(81)
```

```python
# tampilkan tipe data dari y_tf (Tensorflow)
print(f'type dari y_tf: {type(y_tf)}')
```

```python
# jumlahkan x_tf + y_tf (Tensorflow)
jumlah_tf = x_tf + y_tf
print(jumlah_tf)
```

```python
# buat variabel bernama x_float_tf dan simpan suatu nilai desimal (Tensorflow)
x_float_tf = tf.Variable(1.90)
```

```python
# jumlahkan x_float_tf dengan y_tf (Tensorflow)
y_float_tf = float(y_tf)
jumlah_tf = x_float_tf + y_float_tf
print(jumlah_tf)
```

```python
# tampilkan tipe data dari x_float_tf (Tensorflow)
print(f'type dari x_float_tf: {type(x_float_tf)}')
```

```python
"""**NO 3 (SCALAR) (PYTORCH)**"""
```

```python
# lakukan tahapan pada soal 1 dengan menggunakan pytorch
import torch
```

```python
x_pt = torch.tensor(19)
```

```python
# tampilkan tipe data dari x_pt (PyTorch)
print(f'type dari x_pt: {type(x_pt)} nilainya {x_pt}')
```

```python
y_pt = torch.tensor(81)
```

```python
print(f'type dari y_pt: {type(y_pt)} nilainya {y_pt}')
```

```python
jumlah_tf = x_pt + y_pt
print(jumlah_tf)
```

```python
x_float_pt = torch.tensor(1.90)
```

```python
jumlah_tf = x_float_pt + y_pt
print(jumlah_tf)
```

```python
print(f'type dari x_float_pt: {type(x_float_pt)} nilainya {x_float_pt}')
```
