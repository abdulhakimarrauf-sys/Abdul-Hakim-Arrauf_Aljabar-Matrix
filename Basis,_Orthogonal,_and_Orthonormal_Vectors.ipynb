```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
i


j = np.array([0, 1])
j
```

```python
# buktikan bahwa vector i dan vector j itu orthogonal
bagian1 = np.dot(i, j)
if bagian1 == 0:
  print(f'i dan j adalah orthogonal karena matematisnya (1, 0) . (0, 1) = (1 x 0) + (0 x 1) = {bagian1}')
else:
  print(f'i dan j bukan orthogonal karena hasil akhirnya {bagian1}')
```

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
w = np.array([2, -1])
bagian2 = np.dot(v, w)
if bagian2 == 0:
  print(f'v dan w orthogonal, karena hasilnya: {bagian2}')
else:
  print(f'v dan w bukan orthogonal karena hasilnya: {bagian2}')
```
